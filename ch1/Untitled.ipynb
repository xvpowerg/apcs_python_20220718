# Python 螢幕輸出


print() 函式參數進階設定

sep = ' … ' ：設定資料間的分隔符號

end = ' … ' ：設定句尾符號

輸出後不想自動換行

加上參數 end = '' 


```python
print("Test!!")
print(123)
print(1.141421)
print(4.1415,2.71828)#輸出兩個浮點數
print(4,5,6,sep="-",end="\n\n")
print("Python!")
```
