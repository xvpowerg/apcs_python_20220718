## append 附加 pop彈出


```python
list1 = []
print(list1)

list1.append("Ken")
list1.append("Vivin")
list1.append("Lucy")
list1.append("Join")

print(list1)
v1 = list1.pop() #預設最後加入的會彈出
v1 = list1.pop() #預設最後加入的會彈出
print(v1)
print(list1)

list2 = ["A","B","C"]
print(list2)
v2 = list2.pop(0) #依index彈出
print(v2)
print(list2)
```
