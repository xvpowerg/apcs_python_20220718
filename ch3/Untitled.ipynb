## BMI練習


```python
height = float(input('請輸入身高，單位為公分: '))
weight = int(input('請輸入體重，單位為公斤: '))
bmi = weight/((height/100)**2)
print("bmi: %.2f, 判定結果: " %(bmi), end='')

if bmi > 30:
    print("胖")
elif bmi > 25:
    print("重")
elif bmi >18.5:
    print("正常")
elif bmi > 0:
    print("太輕")
else:
    print("錯誤")
```

## 檢查所輸入的數是否為質數

```python
num = int(input('輸入整數:'))

isPrimeNum = True
for i in range(2, num):
    if(num%i==0):
        isPrimeNum = False
        break

if(isPrimeNum):
    print(num, '是質數')
else:
    print(num, '不是質數')
```
