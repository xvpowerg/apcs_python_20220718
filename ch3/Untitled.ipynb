```python
results = {"David":0, "Amy":0, "Sean":0}
candidates = ("David", "Amy", "Sean")

for i in range(5):
    v = input(": ")
    if v not in candidates :
        print("")
        continue 
    results[v] = results[v] + 1

for name in  results:
    print(name,results[name])    
```
