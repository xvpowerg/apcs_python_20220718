## 英文字母接龍

```python
str_result=input('字母接龍\n請輸入字串:')
eCount=0
while(eCount<5):    
    print('目前字串:%s' %str_result)
    str_in = input('請輸入-%s-開始的字串:' %str_result[-1].upper())
    if(str_in[0].lower()==str_result[-1].lower()):
        str_result+="-"+str_in
    else:
        eCount+=1
        print('輸入錯誤%d次' %eCount)

print('錯誤超過五次,遊戲結束!')
```

# 作業：英文單字出現次數

```python
song = """Are you sleeping, are you sleeping, Brother John, Brother John?
Morning bells are ringing, morning bells are ringing.
Ding ding dong, Ding ding dong."""
dict = {}                           # 空字典未來儲存單字計數結果
print("原始歌詞")
print(song)

# 將歌曲大寫字母改成小寫
songLower = song.lower()            # 歌曲改為小寫
print("小寫歌詞")
print(songLower)

# 去掉標點符號
for ch in songLower:                
    if ch in ".,?":                 # 將歌曲的標點符號用空字元取代
        songLower = songLower.replace(ch,'')
print("去掉標點符號")    
print(songLower)

# 將歌曲字串轉成串列
songList = songLower.split()        
print("轉換成串列")
print(songList)                     # 列印歌曲串列

# 將歌曲串列處理成字典 
for wd in songList:                 
    if wd in dict:              # 檢查此字是否已在字典內
        dict[wd] += 1           # 累計出現次數
    else:
        dict[wd] = 1            # 第一次出現的字建立此鍵與值
    
print("單字及出現次數")
print(dict)                         # 列印字典
```

```python
#寫一組函式testScore可以判斷是否及格
# >= 60回傳及格 否則不及格
#
class Student:
    def __init__(self,name,height,score=0):
        self.name = name
        self.height = height
        self.score = score
    def printInfo(self):
        print(self.name,self.height,self.score)  
    def testScore(self):
        if self.score >= 60:
            return "及格"
        else:
            return "不及格"
        
st1 = Student("Ken",175.2)
st1.printInfo()
print(st1.testScore())
# print(st1.name)        
# print(st1.height)
# print(st1.score)
st2 = Student("Vivin",156,75)
st2.printInfo()
print(st2.testScore())
```

```python
class Point:
    def __init__(self,x=0,y=0):
        self.x = x
        self.y = y
    def __str__(self):
        return f"(x:{self.x},y:{self.y})"
    def __add__(self,other):
        x = self.x + other.x
        y = self.y + other.y
        return Point(x,y)
p1 = Point(2,3)
p2 = Point(-1,2)
print(p1)
print(p2)    
print(p1 + p2)   
```

```python
class Point:
    def __init__(self,x=0,y=0):
        self.x = x
        self.y = y
    def __str__(self):
        return f"({self.x},{self.y})"    
    def __lt__(self,other):
        s1 = (self.x ** 2)+(self.y ** 2)
        s2 = (other.x ** 2) + (other.y ** 2)
        return s1 < s2    
p1 = Point(1,1)
p2 = Point(-2,-3)
print(p1)
print(p2)
print(p1 < p2)
```

# Sort

```python
class Student:
    def __init__(self,name,score):
        self.name = name
        self.score = score
    def __str__(self):
        return f"{self.name}:{self.score}"
    def __lt__(self,other):
        return self.score < other.score 
stList = []  
      
st1 = Student("Ken",65) 
st2 = Student("Iris",85)
st3 = Student("Vivin",63)
st4 = Student("Ben",25)
st5 = Student("Lucy",73)
stList.append(st1)
stList.append(st2)
stList.append(st3)
stList.append(st4)
stList.append(st5)
stList.sort(reverse=True)
for s in stList:
    print(s)
```
